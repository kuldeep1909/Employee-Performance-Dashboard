```python
# the distribution of employee ages?
df['Age'].plot.hist()
plt.xlabel('Age')
plt.title("Age Distribution")
```

```python
df.columns
```

```python
# How does the distribution of ages vary by gender?
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Age', data=df)
plt.title('Distribution of Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# What are the proportions of different education backgrounds?

df['EducationBackground'].value_counts()
```

```python
df['EducationBackground'].value_counts().plot.pie( )
```

```python
# Job Satisfaction and Performance

sns.jointplot(x='EmpJobSatisfaction', y='PerformanceRating', data=df, kind='reg', height=7)
plt.suptitle('Job Satisfaction vs Employee Performance Joint Plot', y=1.02)
plt.show()
```

```python
# the relationship between job satisfaction and performance ratings?

pd.crosstab(df['EmpJobSatisfaction'], df['PerformanceRating']).plot(kind='bar')
```

```python
# How does job involvement vary with job satisfaction?

pd.crosstab(df['EmpJobInvolvement'], df['EmpJobSatisfaction']).plot.bar()
```

```python
# correlation between job satisfaction and other features?
for column in ['EmpEducationLevel', 'EmpEnvironmentSatisfaction', 'EmpJobInvolvement', 
               'EmpJobSatisfaction', 'PerformanceRating', 'EmpRelationshipSatisfaction', 
               'EmpWorkLifeBalance']:
    plt.figure(figsize=(10, 6))
    df[column].value_counts().plot(kind='bar', color=[ 'salmon'])
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()

    print(f"\nPercentage distribution of {column}:")
    print(df[column].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='PerformanceRating', data=df)
plt.title('Distribution of Performance Ratings')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='EmpJobSatisfaction', data=df)
plt.title('Distribution of Job Satisfaction')
plt.show()

for column in ['EmpEducationLevel', 'EmpEnvironmentSatisfaction', 'EmpJobInvolvement', 
               'EmpRelationshipSatisfaction', 'EmpWorkLifeBalance']:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=column, y='EmpJobSatisfaction', data=df)
    plt.title(f'{column} vs Job Satisfaction')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='EmpWorkLifeBalance', data=df)
plt.title('Distribution of Work-Life Balance')
plt.show()

sns.pairplot(df, vars=['EmpWorkLifeBalance', 'EmpJobSatisfaction', 'PerformanceRating'])
plt.suptitle('Pairplot of Work-Life Balance, Job Satisfaction, and Performance Rating', y=1.02)
plt.show()
```
